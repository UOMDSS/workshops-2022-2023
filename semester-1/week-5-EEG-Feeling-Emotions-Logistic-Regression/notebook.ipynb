```python
import torch, pandas
import matplotlib.pyplot as plt
```

```python
data = pandas.read_csv("data/emotions.csv")
data.head()
```

```python
print(data.shape)
data.describe()
```

```python
for (i, col) in enumerate(data.columns):
    print(i, col)
```

```python
plt.figure(figsize=(10, 20))
plt.plot(data.columns[:-1], data.iloc[0, :-1])
```

```python
ranges = [("fft_0_a", "fft_479_a"), ("label", "label")]
fft_data = pandas.concat([data.loc[:, i:j] for i, j in ranges], axis=1)[
    data["label"] != "NEUTRAL"].reset_index()
```

```python
fft_data
```

```python

fft_data.iloc[0, 1:-1].plot(figsize=(20, 15), label="Fast Fourier Transform at interval 0s to 1s")
plt.legend()
```

```python
# make multiple plots
fig = plt.figure(figsize=(25, 15))
for i in range(0, 12):
    plt.subplot(4, 3, i+1)
    plt.plot(fft_data.iloc[i*100, 1:-1])
    img = plt.imread("assets/happy.png") if fft_data.iloc[i*100, -1] == "POSITIVE" else plt.imread("assets/sad.png")
    plt.title("Fast Fourier Transform on " + fft_data.loc[i*100, "label"] + " emotion")
```
